```python
import pandas as pd
from sklearn.linear_model import LinearRegression,BayesianRidge,LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
from sklearn.neighbors import KNeighborsRegressor,KNeighborsClassifier
from sklearn.ensemble import GradientBoostingRegressor,GradientBoostingClassifier
from sklearn.svm import SVR,SVC
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as py
import seaborn as sb
sb.set(rc={"figure.figsize":(10,5)})
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stop = stopwords.words('english')
from sklearn.feature_extraction.text import CountVectorizer
c=CountVectorizer()
from nltk.stem import PorterStemmer,WordNetLemmatizer
s=PorterStemmer()
l=WordNetLemmatizer()
```

```python
t=pd.read_csv("D:/Datasets/Twitter sentiment analysis/train_E6oV3lV.csv")
```

```python
t
```

```python
t["tweet"][1]
```

```python
t.drop("id",axis=1,inplace=True)
```

```python
t
```

```python
t.label.value_counts()
```

```python
t["tweet"]=t["tweet"].apply(lambda x: l.lemmatize(x))
```

```python
t["tweet"]=t["tweet"].apply(lambda x: " ".join([i for i in x.split() if i.startswith("#")==False and i.startswith("@")==False ]))
```

```python
t["tweet"]=t["tweet"].apply(lambda x:word_tokenize(x))
```

```python
t["tweet"]
```

```python
t["tweet"]=t["tweet"].apply(lambda x: [i for i in x if i.isalpha()])
```

```python
t["tweet"]=t["tweet"].apply(lambda x: [i for i in x if i not in stop])
```

```python
t["tweet"]=t["tweet"].apply(lambda x: [i.lower() for i in x])
```

```python
t["tweet"]=t["tweet"].apply(lambda x: " ".join([i for i in x]))
```

```python
mat=pd.DataFrame(c.fit_transform(t["tweet"]).toarray(),columns=c.get_feature_names(),index=None)
mat
```

```python
cross_val_score(LogisticRegression(),mat,t["label"],cv=10).mean()
```

```python
from sklearn.naive_bayes import MultinomialNB
cross_val_score(MultinomialNB(),mat,t["label"],cv=10).mean()
```
